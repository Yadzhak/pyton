Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: 
surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv', sep=',', header = None, 
                 names= ['surgery?', 'Age', 'Hospital Number', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 
                         'peripheral pulse', 'mucous membranes', 'capillary refill time', 'pain', 'peristalsis', 'abdominal distension', 
                         'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination', 'abdomen', 
                         'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 
                         'surgical lesion?', '25 type of lesion', '26 type of lesion', '27 type of lesion', 'cp_data'], na_values = '?')

df.head()
```

```python
horse_database = df[['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']]
horse_database.loc[(horse_database.Age == 9), 'Age'] = 2 #Заменим значение "9" на "2" в столбце "Age"
horse_database
```

Задание 2. Первичное изучение данных
Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
horse_database.describe()
```

  1:  surgery?
          1 = Yes, it had surgery
          2 = It was treated without surgery

  2:  Age
          1 = Adult horse
          2 = Young (< 6 months)

  4:  rectal temperature
          - linear
          - in degrees celsius.
          - An elevated temp may occur due to infection.
          - temperature may be reduced when the animal is in late shock
          - normal temp is 37.8
          - this parameter will usually change as the problem progresses
               eg. may start out normal, then become elevated because of
                   the lesion, passing back through the normal range as the
                   horse goes into shock

  5:  pulse
          - linear
          - the heart rate in beats per minute
          - is a reflection of the heart condition: 30 -40 is normal for adults
          - rare to have a lower than normal rate although athletic horses
            may have a rate of 20-25
          - animals with painful lesions or suffering from circulatory shock
            may have an elevated heart rate

   6:  respiratory rate
          - linear
          - normal rate is 8 to 10
          - usefulness is doubtful due to the great fluctuations

  7:  temperature of extremities
          - a subjective indication of peripheral circulation
          - possible values:
               1 = Normal
               2 = Warm
               3 = Cool
               4 = Cold
          - cool to cold extremities indicate possible shock
          - hot extremities should correlate with an elevated rectal temp.

11: pain - a subjective judgement of the horse's pain level
          - possible values:
               1 = alert, no pain
               2 = depressed
               3 = intermittent mild pain
               4 = intermittent severe pain
               5 = continuous severe pain
          - should NOT be treated as a ordered or discrete variable!
          - In general, the more painful, the more likely it is to require
            surgery
          - prior treatment of pain may mask the pain level to some extent

23: outcome
          - what eventually happened to the horse?
          - possible values:
               1 = lived
               2 = died
               3 = was euthanized
  
  Качественные данные - surgery?,	Age,	temperature of extremities,	pain,	outcome. Они не подлежат сравнению. Судя по выводу функции describe() с учетом описания базы данных, выбросы в них отсутствуют.

```python
print('Основные показатели:', horse_database[['rectal temperature', 'pulse', 'respiratory rate']].describe(),sep='\n', end='\n\n')
print('Мода:', horse_database[['rectal temperature', 'pulse', 'respiratory rate']].round().mode(), sep='\n', end='\n\n')
print('Медиана:', horse_database[['rectal temperature', 'pulse', 'respiratory rate']].median(), sep='\n', end='\n\n')
print('Дисперсия:', horse_database[['rectal temperature', 'pulse', 'respiratory rate']].var(), sep='\n', end='\n\n')
```

## Находим выбросы

```python
q1_rt = horse_database['rectal temperature'].quantile(0.25)
q3_rt = horse_database['rectal temperature'].quantile(0.75)
iqr_rt = q3_rt - q1_rt
lower_bound_rt = q1_rt - (1.5 * iqr_rt) 
upper_bound_rt = q3_rt + (1.5 * iqr_rt)
remove_outliers_rt = horse_database[horse_database['rectal temperature'].between(lower_bound_rt, upper_bound_rt, inclusive=True)].sort_values('rectal temperature')
rt_table = pd.concat([horse_database, remove_outliers_rt]).drop_duplicates(keep=False)
list(rt_table['rectal temperature'].dropna())
```

```python
q1_p = horse_database['pulse'].quantile(0.25)
q3_p = horse_database['pulse'].quantile(0.75)
iqr_p = q3_p - q1_p
lower_bound_p = q1_p - (1.5 * iqr_p) 
upper_bound_p = q3_p + (1.5 * iqr_p)
remove_outliers_p = horse_database[horse_database['pulse'].between(lower_bound_p, upper_bound_p, inclusive=True)].sort_values('pulse')
p_table = pd.concat([horse_database, remove_outliers_p]).drop_duplicates(keep=False)
list(p_table['pulse'].dropna())
```

```python
q1_rr = horse_database['respiratory rate'].quantile(0.25)
q3_rr = horse_database['respiratory rate'].quantile(0.75)
iqr_rr = q3_rr - q1_rr
lower_bound_rr = q1_rr - (1.5 * iqr_rr) 
upper_bound_rr = q3_rr + (1.5 * iqr_rr)
remove_outliers_rr = horse_database[horse_database['respiratory rate'].between(lower_bound_rr, upper_bound_rr, inclusive=True)].sort_values('respiratory rate')
rr_table = pd.concat([horse_database, remove_outliers_rr]).drop_duplicates(keep=False)
list(rr_table['respiratory rate'].dropna())
```

В данном примере выбросы не являются аномальными, будем считать их вариантом нормы. Поэтому они не требуют удаления.

Создаем датафрейм для заполнения пропусков:

```python
test_fill = df[['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome']]
test_fill.loc[(horse_database.Age == 9), 'Age'] = 2
```

```python
test_fill.describe()
```

```python
test_fill['respiratory rate'].fillna(test_fill['respiratory rate'].mean(), inplace=True)
test_fill['pulse'].fillna(test_fill['pulse'].mean(), inplace=True)
test_fill['rectal temperature'].fillna(test_fill['rectal temperature'].mean(), inplace=True)
test_fill.describe()
# Заполнил количественные данные средними значениями, при этом статистические показатели изменились незначительно
```

```python
print(test_fill.groupby(['surgery?', 'outcome'])['pain'].median()) 
```

```python
print(test_fill.groupby('pain')['temperature of extremities'].median())
```

```python
test_fill['pain'].fillna(test_fill.groupby(['surgery?', 'outcome'])['pain'].transform('median'), inplace=True)
test_fill['temperature of extremities'].fillna(test_fill.groupby('pain')['temperature of extremities']
                                               .transform('median'), inplace=True)

# Столбец 'pain'  в зависимости от того, была ли операция у лошади и каков исход(жива/мертва/усыплена)
# Столбец 'temperature of extremities' кореллирует в высокой температурой тела, которая зависит от боли.
# Поэтому заполнил медианным значением с группировкой по уровню боли.
```
