```python
#перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
    #перечень полок, на которых хранятся документы 
    #(если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Находим владельца

def find_owner():
    num = input('Введите номер документа: ')
    for line in documents:
        if num == line['number']:
            return print('Владелец документа: ', line['name'])
                   
    print('Документ не найден в базе')
# find_owner()

# Находим полку

def find_shelf():
    num = input('Введите номер документа: ')
    for key, value in directories.items():
        if num in value:
            return print('Документ хранится на полке:', key)
                   
    print('Документ не найден в базе')
# find_shelf()

# Информация по документам

def find_info():
    for line in documents:
        for key, value in directories.items():
            if line['number'] in value:
                print('Номер документа:', line['number'], ',' , 'тип: ', line['type'], ',' ,'владелец: ', line['name'], ',', 'полка хранения: ', key[0])
    return
# find_info()

# Добавляем полку

def add_shelf():
    num = input('Введите номер полки: ')
    if num in directories.keys():
        print('Такая полка уже существует. Текущий перечень полок:', list(directories))
    
    else:
        directories[num] = []
        print('Полка добавлена. Текущий перечень полок:', list(directories))
    return
# add_shelf()    

# Удаляем полку

def del_shelf():
    shelf = input('Введите номер полки: ')
    if shelf in directories.keys():
        if len(directories[shelf]) == 0:
            del(directories[shelf])
            print('Полка удалена. Текущий перечень полок: ', list(directories))
        else:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', list(directories))
    else:
        print('Такой полки не существует. Текущий перечень полок: ', list(directories))
    return
# del_shelf()   

# Основная функция
def main():
    input_cmd = '\n Введите команду: '
    while True:
        cmd = input(input_cmd)
        if cmd == 'p':
            find_owner()

        elif cmd == 's':
            find_shelf()
                
        elif cmd == 'l':
            find_info()
                
        elif cmd == 'ads':
            add_shelf()
                
        elif cmd == 'ds':
            del_shelf()
                
        elif cmd == 'q':
            break
    else:
        print('Неверная команда')
                  
main()
```
