```python
# Задание 1

import pandas as pd
df = pd.read_csv('ratings.csv')
df.info() # Проверяю, что значения в столбцах не нулевые

def ratings(value):
    
    if value <= 2.0 :
        return 'Low'
            
    elif value > 2.0 and value <= 4.0 :
        return 'Average'
        
    elif value == 5.0 or value == 4.5:
        return 'High'
    
df['class'] = df['rating'].apply(ratings)
df.head(50)
```

```python
# Задание 2
import pandas as pd
kw = pd.read_csv('keywords.csv')

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def classify(kw):
    for key, value in geo_data.items():
        for vals in value:
            if vals in kw['keyword']:
                return key
    return 'undefined'

kw['region'] = kw.apply(classify, axis = 1)
list(kw['region'])
```

```python
# Задание 3
import pandas as pd
df_ratings = pd.read_csv('ratings.csv')
df_movies = pd.read_csv('movies.csv')

years_list = list(range(1950,2011))

def production_year(df_movies):
  for year in years_list:
    if str(year) in df_movies['title']:
      return str(year)
    
  return '1900'

df_movies['year'] = df_movies.apply(production_year, axis=1)

a = df_movies[['movieId', 'year']]
b = df_ratings[['movieId', 'rating']]
c = b.merge(a, left_on='movieId', right_on='movieId')
avg_rating = (c[['rating', 'year']].groupby('year').sum() / c[['rating', 'year']].groupby('year').count()).sort_values(by = 'rating', ascending=False)
avg_rating
```
